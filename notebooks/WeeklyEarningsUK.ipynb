```python

earnings = pd.read_excel('../data/realAverageWeeklyEarnings.xlsx', sheet_name='AWE Real_CPI', header=[5,6], nrows=282)

earnings
```

```python
earnings = earnings.rename(columns={'Unnamed: 0_level_0': 'Date'})
earnings
```

```python
formattedTotalEarnings = pd.DataFrame()
formattedTotalEarnings['Date'] = earnings['Date']
formattedTotalEarnings['Real AWE'] = earnings['Real AWE\n(2015 Â£)', 'Unnamed: 5_level_1']
formattedTotalEarnings['Percentage change year on year (single month)'] = earnings['% changes year on year', 'Single month.1']
formattedTotalEarnings['Percentage change year on year (3 month average month)'] = earnings['% changes year on year', '3 month average.1']

formattedTotalEarnings = formattedTotalEarnings.dropna()
formattedTotalEarnings
```

```python
formattedTotalEarnings.to_excel('../data/processed/realAverageWeeklyTotalEarnings.xlsx')
```
